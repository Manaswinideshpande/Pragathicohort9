Histogram

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

# Basic histogram
sns.histplot(
    data=df,
    x="SalePrice",       # numeric variable for histogram
    bins=10,             # number of bins
    kde=True,            # add kernel density curve
    color="blue"      
    )

plt.title("Histogram of Sale Prices")
plt.xlabel("SalePrice")
plt.ylabel("Count")

plt.show()
```

***observation***
The histogram indicates that property sales gradually increase from $200,000 to around $350,000, show a slight dip around $400,000–$420,000, then rise again and peak near $520,000 before declining sharply beyond $530,000, with fewer transactions in the highest price range.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

# Save the scatterplot object
scatter = sns.scatterplot(
    data=df,
    x="Location",          # categorical axis
    y="SalePrice",         # numeric axis
    hue="Size_sqft",       # continuous variable for color
    palette="viridis",     # gradient palette
    size="Size_sqft",      # optional: scale point sizes
    sizes=(50, 200),
    legend=False           # disable default legend
)

plt.title("Scatter Plot")
plt.ylabel("SalePrice")
plt.xlabel("Location")

# Attach colorbar to the scatterplot
norm = plt.Normalize(df["Size_sqft"].min(), df["Size_sqft"].max())
sm = plt.cm.ScalarMappable(cmap="viridis", norm=norm)
sm.set_array([])

plt.colorbar(sm, ax=scatter.axes, label="Size_sqft")

plt.show()
```

***Observation***:
1.In the Suburb location, higher sale prices are generally associated with medium-to-large property sizes, suggesting a positive relationship between property area and market value. However, the relationship is not perfectly linear.

2.In the Downtown location, sale prices increase with property size up to approximately $350,000. Between $350,000 and $470,000, properties of varying sizes (approximately 1200–2000 sqft) appear within a similar price range, indicating a non-linear relationship between size and price. Beyond $470,000, larger properties again tend to correspond with higher sale prices. An outlier is observed near $600,000 for a comparatively smaller property, suggesting that additional factors such as property age, premium amenities, interior quality, neighborhood safety, proximity to public transport, and market fluctuations may significantly influence pricing.

3.In the Countryside location, most properties sized between 2000 and 2500 sqft are sold within the $230,000–$450,000 range. A notable outlier exists where a relatively smaller property (~1500 sqft) is priced at approximately $600,000, indicating the influence of factors beyond property size.

```python
# correlation between Location and sales
# correlation between size_sqft and sales

# encoding categorical data
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df["Location_encoded"] = le.fit_transform(df["Location"])

print(df[["Location", "Location_encoded"]].head())
```

```python
#pivot table
df["Size_bin"] = pd.cut(df["Size_sqft"], bins=10)

pivot_data = df.pivot_table(index="Size_bin",
                            columns="Location",
                            values="SalePrice",
                            aggfunc="mean")

pivot_data.plot(kind="bar", figsize=(18,8))

plt.title("SalePrice vs Size bins by Location")
plt.ylabel("Avg SalePrice")
plt.tight_layout()
plt.show()
```

observations:
From the pivot table, it is observed that properties with a size range of 803–1025 sq. ft. have higher sales in the Countryside location and comparatively lower sales in the Downtown area.

***Correlation***

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# ANOVA model
model = ols('SalePrice ~ C(Location)', data=df).fit()

anova_table = sm.stats.anova_lm(model, typ=2)

print(anova_table)
```

Conclusion:
From the above EDA analysis it is observed that the location has no significant effect on saleprice 
also area size in sqfeet of house maintains non linear relationship with saleprice and 
The ANOVA test shows that Location does not have a statistically significant effect on SalePrice (F = 2.01, p = 0.139 > 0.05). 
Therefore, we fail to reject the null hypothesis and conclude that mean sale prices are not significantly different across locations.
