### Implementation and Analysis of Enhanced Movie Recommendation Engine

To enhance the performance of the movie recommendation engine, the analysis was broadened by incorporating additional movie features beyond the basic movie title. #### Steps Followed

1. **Dataset Loading**
   The movie dataset was successfully loaded and inspected to understand its structure, feature distribution, and missing values.

2. **Handling Missing Values**
   Missing values present in the dataset were handled using appropriate techniques such as filling null values with suitable replacements or removing highly irrelevant columns with excessive missing data.

3. **Data Preprocessing and Feature Engineering**
   Preprocessing was performed using the `extract_names` function to extract meaningful textual information from structured columns such as genres. Additionally, a custom `preprocess` function was used to clean the textual data by removing special characters, converting text to lowercase, removing stop words, and performing lemmatization. This produced clean text data suitable for the next stage of vectorization.

4. **Vectorization using TF-IDF**
   The cleaned text data was converted into vector embeddings using the TF-IDF (Term Frequency–Inverse Document Frequency) vectorization technique. TF-IDF helps capture the importance of words within movie descriptions and metadata by assigning higher importance to informative words and lower importance to frequently occurring common words. Since the objective of this project is to group and recommend similar movies based on their content, TF-IDF proved to be an appropriate vectorization method.

5. **Recommendation System Implementation**
   A recommendation function was implemented to generate movie recommendations based on user-provided movie titles. The system identifies movies with similar content using cosine similarity, a distance-based similarity metric that measures how closely related movie vectors are in vector space.

#### Impact of Enhancements

The inclusion of additional movie features improved the recommendation engine by providing richer contextual information about each movie. Features such as genres and descriptive metadata enabled the system to generate more relevant recommendations by identifying stronger semantic similarities between movies.

The enhanced recommendation engine improved the overall user experience by recommending movies that closely matched user interests and selected movie preferences. The cosine similarity–based approach successfully identified neighboring movie vectors and returned highly relevant movie recommendations based on the chosen features.

```python
import ast

def extract_names(text):
    if pd.isna(text) or text == '':
        return ''
    
    try:
        data = ast.literal_eval(text)
        return " ".join([i['name'] for i in data])
    except:
        return ''
```

```python
import nltk

nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

lemma = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess(data):
    
    # remove special characters
    sentence = re.sub("[^a-zA-Z]", " ", str(data))
    
    # lowercase
    sentence = sentence.lower()
    
    # tokenization
    sentence = sentence.split()
    
    # stopword removal
    sentence = [word for word in sentence if word not in stop_words]
    
    # lemmatization
    sentence = [lemma.lemmatize(word) for word in sentence]
    
    return " ".join(sentence)
```

```python
cleaned_string2 = [preprocess(doc) for doc in string2]
print(cleaned_string2)
```
